```python
import os
import glob
import pandas as pd
from shutil import copyfile, copy

folders = [x[0] for x in os.walk('./RawData/')]

for folder in folders:
    sub_id = folder.split("/")[2][:6]
    images = glob.glob(os.path.join(folder, "*.jpg"))
    images.append(glob.glob(os.path.join(folder, "*.png")))
    
    if len(images) > 1:
        destination = os.path.join("./data", sub_id)
        if not os.path.exists(destination):
            os.mkdir(destination)

        for source in images:
            if len(source) > 0:
                try:
                    copy(str(source), str(destination))
                except:
                    print(f"Error occur: {source}")
```

```python
folders = [x[0] for x in os.walk('./data/')]
for folder in folders:
    sub_id = folder.split("/")[2][:6]
    images = glob.glob(os.path.join(folder, "*.jpg"))
    images.append(glob.glob(os.path.join(folder, "*.png")))

    for index, source in enumerate(images):
        if len(source) > 0:
            path = source.split("/")
            fname = path[-1]
            extend = fname.split(".")[-1]

            new_fname = sub_id + "_" + str(index + 1).zfill(4) + "." + extend
            new_fpath = os.path.join("/".join(path[:-1]), new_fname)
            os.rename(source, new_fpath)
```
